# **Topic of OEP : Weight Condition Prediction**
#### This helps to predict whether a person's weight is optimal, overweight, obese, or severely obese.

### **Information About DataSet**

  This dataset helps estimate obesity levels based on eating habits, family history and physical condition. It includes data from individuals in Mexico, Peru, and Colombia, covering 16 lifestyle and health-related features with 2111 records. The labels classify obesity levels, ranging from underweight to different obesity types.

  Most of the data was generated using synthetic techniques, while some were collected directly from users through a web platform. It is useful for classification, regression, and clustering tasks.

### **Source**:
UCI Machine Learning Repository

### **Inspiration**:
Understanding lifestyle factors that contribute to obesity can help with early intervention, health recommendations, and machine learning applications in healthcare.

### **Column Descriptions**:

**Gender** – Male or Female.

**Age** – The person’s age in years.

**Height** – Height in meters.

**Weight** – Weight in kilograms.

**family_history_with_overweight** – Whether the person has a family history of being overweight (yes/no).

**FAVC** – If the person frequently consumes high-calorie foods (yes/no).

**FCVC** – Frequency of vegetable consumption (scale from 1 to 3).

**NCP** – Number of main meals per day.

**CAEC** – Frequency of consuming food between meals (Never, Sometimes, Frequently, Always).

**SMOKE** – Whether the person smokes (yes/no).

**CH2O** – Daily water intake (scale from 1 to 3).

**SCC** – If the person monitors their calorie intake (yes/no).

**FAF** – Physical activity frequency (scale from 0 to 3).

**TUE** – Time spent using technology (scale from 0 to 3).

**CALC** – Frequency of alcohol consumption (Never, Sometimes, Frequently, Always).

**MTRANS** – Main mode of transportation (Automobile, Bike, Motorbike, Public Transportation, Walking).

**NObeyesdad** – Obesity level (Insufficient Weight, Normal Weight, Overweight Level I, Overweight Level II, Obesity Type I, Obesity Type II, Obesity Type III).

This dataset provides a structured view of how different lifestyle factors relate to obesity levels, making it useful for healthcare analysis and decision-making.

## **Step 1 : Importing Libraries & DataSet**

```python
import pandas as pd
import numpy as np
data = pd.read_csv("Obesity.csv")
data.sample(10)
```

## **Step 2 : Handling Null Values**

```python
maps = data.isnull().sum()
maps
```

### Inference : No need to handle missing value since there is none.

## **Step 3.1 : Adding a row - BMI, converts two features(Height and Weight) into one(BMI), to reduce dimensionality.**

```python
data['BMI'] = data.Weight / (data.Height)**2
data.BMI
```

## **Step 3.2 : Creating a column that shows whether a person is obese or not.**

```python
obesity_enc = []
for i in range(data.shape[0]):
    if data.NObeyesdad.iloc[i] == 'Obesity_Type_I' or data.NObeyesdad.iloc[i] == 'Obesity_Type_II' or data.NObeyesdad.iloc[i] == 'Obesity_Type_III':
        obesity_enc.append(1)
    else:
        obesity_enc.append(0)

data['obesity_enc'] = obesity_enc
```

## **Step 4.1 : Encoding family history with overweight with 0 - NO and 1 - YES.**

```python
data['FHWO_Enc'] = [1 if data.family_history_with_overweight.iloc[i] == 'yes' else 0 for i in range(data.shape[0])]
data['FHWO_Enc']
```

## **Step 4.2 : Encoding If the person frequently consumes high-calorie foods (yes/no) as (1/0)**

```python
data['FAVC_Enc'] = [1 if data.FAVC.iloc[i] == 'yes' else 0 for i in range(data.shape[0])]
data['FAVC_Enc']
```

## **Step 4.3 Encoding Frequency of consuming food between meals (Never, Sometimes, Frequently, Always) as (0/1/2/3)**

```python
CAEC_Enc = []
for i in range(data.shape[0]):
    if data.CAEC.iloc[i] == 'no':
        CAEC_Enc.append(0)
    elif data.CAEC.iloc[i] == 'Sometimes':
        CAEC_Enc.append(1)
    elif data.CAEC.iloc[i] == 'Frequently':
        CAEC_Enc.append(2)
    elif data.CAEC.iloc[i] == 'Always':
        CAEC_Enc.append(3)

print(data['CAEC'].value_counts())
print(len(CAEC_Enc))
data['CAEC_Enc'] = CAEC_Enc
```

## **Step 4.4 : CALC Encoding**

```python
CALC_Enc = []
for i in range(data.shape[0]):
    if data.CALC.iloc[i] == 'no':
        CALC_Enc.append(0)
    elif data.CALC.iloc[i] == 'Sometimes':
        CALC_Enc.append(1)
    elif data.CALC.iloc[i] == 'Frequently':
        CALC_Enc.append(2)
    elif data.CALC.iloc[i] == 'Always':
        CALC_Enc.append(3)

print(data['CALC'].value_counts())
print(len(CALC_Enc))
data['CALC_Enc'] = CALC_Enc
```

## **Step 4.5 : Encoding SCC values to 0/1 for no/yes.**

```python
data['SCC_Enc'] = [1 if data.SCC.iloc[i] == 'yes' else 0 for i in range(data.shape[0])]
data['SCC_Enc']
```

## **Step 4.6 : Encoding Target field in the range of 0-7 for Insufficient Weight, Normal Weight, Overweight Level I, Overweight Level II, Obesity Type I, Obesity Type II, Obesity Type III, respectively.**

```python
NObeyesdad_Enc = []
for i in range(data.shape[0]):
    if data.NObeyesdad.iloc[i] == 'Insufficient_Weight':
        NObeyesdad_Enc.append(0)
    elif data.NObeyesdad.iloc[i] == 'Normal_Weight':
        NObeyesdad_Enc.append(1)
    elif data.NObeyesdad.iloc[i] == 'Overweight_Level_I':
        NObeyesdad_Enc.append(2)
    elif data.NObeyesdad.iloc[i] == 'Overweight_Level_II':
        NObeyesdad_Enc.append(3)
    elif data.NObeyesdad.iloc[i] == 'Obesity_Type_I':
        NObeyesdad_Enc.append(4)
    elif data.NObeyesdad.iloc[i] == 'Obesity_Type_II':
        NObeyesdad_Enc.append(5)
    elif data.NObeyesdad.iloc[i] == 'Obesity_Type_III':
        NObeyesdad_Enc.append(6)

print(data['NObeyesdad'].value_counts())
print(len(NObeyesdad_Enc))
data['NObeyesdad_Enc'] = NObeyesdad_Enc
```

## **Step 4.7 : Gender Encoding to 0/1 for male/female**

```python
data['Gender_Enc'] = [1 if data.Gender.iloc[i] == 'Female' else 0 for i in range(data.shape[0])]
data['Gender_Enc']
```

## **Step 5.1 : Visual analysis for outliers**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data.Height)
plt.title('Height analysis')
plt.show()

sns.boxplot(data.Weight)
plt.title('Weight analysis')
plt.show()

sns.boxplot(data.Age)
plt.title('Age analysis')
plt.show()

sns.boxplot(data.BMI)
plt.title('BMI analysis')
plt.show()
```

## **Step 5.2 : Removing weight outlier**

```python
import numpy as np
q1 = np.quantile(data.Weight, 0.25)
q3 = np.quantile(data.Weight, 0.75)
iqr = q3-q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
data = data[(data.Weight >= lower_bound) & (data.Weight <= upper_bound)]
data.shape
```

##**Step 5.3 : Reducing Age outliers. Since there is multiple outliers in age, we can't completely remove it. So we'll limit our age range to some point.**

```python
data.Age.value_counts()
```

```python
data = data[(data.Age >= 16) & (data.Age <= 40)]
data.shape
```

```python
sns.boxplot(data.Age)
plt.show()
sns.boxplot(data.Weight)
plt.show()
```

## **Step 5.4 : Removing the only record that has CALC value 'Always'.**

```python
data = data[data.CALC != 'Always']
data.shape
```

## **Step 6 : Association between different fields to Target fields.**

```python
ct1 = pd.crosstab(data.family_history_with_overweight, data.NObeyesdad)
ob_yes = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])
ob_total = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])  + (ct1["Obesity_Type_I"]['no'] + ct1["Obesity_Type_II"]['no'] + ct1["Obesity_Type_III"]['no'])
ob_pr = ob_yes / ob_total
print(f"Among those who are obese approx {round(ob_pr*100)}% have family history with obesity.")
```

```python
ct1 = pd.crosstab(data.FAVC, data.NObeyesdad)
ob_yes = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])
ob_total = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])  + (ct1["Obesity_Type_I"]['no'] + ct1["Obesity_Type_II"]['no'] + ct1["Obesity_Type_III"]['no'])
ob_pr = ob_yes / ob_total
print(f"Among those who are obese approx {round(ob_pr*100)}% consumes high calorie food frequently.")
```

```python
ct1 = pd.crosstab(data.SCC, data.NObeyesdad)
ob_yes = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])
ob_total = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])  + (ct1["Obesity_Type_I"]['no'] + ct1["Obesity_Type_II"]['no'] + ct1["Obesity_Type_III"]['no'])
ob_pr = ob_yes / ob_total
print(f"Among those who are obese approx {(ob_pr*100)}% keeps track of their calorie intake.")
```

```python
ct2 = pd.crosstab(data.CAEC, data.NObeyesdad)
ob_sometimes = ct2['Obesity_Type_I']['Sometimes'] + ct2['Obesity_Type_II']['Sometimes'] + ct2['Obesity_Type_II']['Sometimes']
ob_no = ct2['Obesity_Type_I']['no'] + ct2['Obesity_Type_II']['no'] + ct2['Obesity_Type_II']['no']
ob_Frequently = ct2['Obesity_Type_I']['Frequently'] + ct2['Obesity_Type_II']['Frequently'] + ct2['Obesity_Type_II']['Frequently']
ob_Always = ct2['Obesity_Type_I']['Always'] + ct2['Obesity_Type_II']['Always'] + ct2['Obesity_Type_II']['Always']
total = ob_sometimes + ob_no + ob_Frequently + ob_Always
pr_sometimes = ob_sometimes / total
pr_no = ob_no / total
pr_Frequently = ob_Frequently / total
pr_Always = ob_Always / total

print(f"Among those who are obese approx {round(pr_sometimes*100, 2)}% consumes foob between meals sometimes.")
print(f"Among those who are obese approx {round(pr_no*100, 2)}% consumes foob between meals never.")
print(f"Among those who are obese approx {round(pr_Frequently*100, 2)}% consumes foob between meals frequently.")
print(f"Among those who are obese approx {round(pr_Always*100, 2)}% consumes foob between meals always.")
```

```python
ct2 = pd.crosstab(data.CALC, data.NObeyesdad)
ob_sometimes = ct2['Obesity_Type_I']['Sometimes'] + ct2['Obesity_Type_II']['Sometimes'] + ct2['Obesity_Type_II']['Sometimes']
ob_no = ct2['Obesity_Type_I']['no'] + ct2['Obesity_Type_II']['no'] + ct2['Obesity_Type_II']['no']
ob_Frequently = ct2['Obesity_Type_I']['Frequently'] + ct2['Obesity_Type_II']['Frequently'] + ct2['Obesity_Type_II']['Frequently']
total = ob_sometimes + ob_no + ob_Frequently
pr_sometimes = ob_sometimes / total
pr_no = ob_no / total
pr_Frequently = ob_Frequently / total

print(f"Among those who are obese approx {round(pr_sometimes*100, 2)}% consumes alcohol sometimes.")
print(f"Among those who are obese approx {round(pr_no*100, 2)}% consumes alcohol never.")
print(f"Among those who are obese approx {round(pr_Frequently*100, 2)}% consumes alcohol frequently.")
```

```python
ct1 = pd.crosstab(data.Gender, data.NObeyesdad)
ob_male = (ct1["Obesity_Type_I"]["Male"] + ct1["Obesity_Type_II"]['Male'] + ct1["Obesity_Type_III"]['Male'])
ob_total = (ct1["Obesity_Type_I"]["Female"] + ct1["Obesity_Type_II"]['Female'] + ct1["Obesity_Type_III"]['Female'])  + (ct1["Obesity_Type_I"]['Male'] + ct1["Obesity_Type_II"]['Male'] + ct1["Obesity_Type_III"]['Male'])
ob_pr = ob_male / ob_total
print(f"Among those who are obese approx {round(ob_pr*100, 2)}% are male.")
print(f"Among those who are obese approx {100 - (round(ob_pr*100, 2))}% are female.")
```

```python
ct1 = pd.crosstab(data.SMOKE, data.NObeyesdad)
ob_yes = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])
ob_total = (ct1["Obesity_Type_I"]["yes"] + ct1["Obesity_Type_II"]['yes'] + ct1["Obesity_Type_III"]['yes'])  + (ct1["Obesity_Type_I"]['no'] + ct1["Obesity_Type_II"]['no'] + ct1["Obesity_Type_III"]['no'])
ob_pr = ob_yes / ob_total
print(f"Among those who are obese approx {(ob_pr*100)}% smokes.")
print(f"Among those who are obese approx {100 - (round(ob_pr*100, 2))}% doesn't smoke.")
```

```python
from sklearn.metrics import accuracy_score
y_pred1 = model1.predict(np.reshape(x_test, (411,1)))
model1.score(np.reshape(x_test, (411,1)), np.reshape(y_test, (411,1)))
```
